```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("G:\math and stat\movie_revenues.csv")
df.head(2)
```

```python
df.revenue.describe()
```

```python
df['revenue_mln'] = df['revenue'].apply(lambda x: x/1000000)
df.revenue_mln.describe()
```

```python
df.head(3)
```

```python
_,mean,std, *_= df.revenue_mln.describe()
```

```python
mean,std
```

```python
def get_zscore(value, mean, std):
    return (value - mean)/std
```

```python
df['z_score'] = df.revenue_mln.apply(lambda x: get_zscore(x, mean, std))
df.head(3)
```

```python
df[df.z_score>3]
```

```python
def get_mad(s):
    median = np.median(s)
    diff = abs(s-median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_mad(df.revenue_mln)
median = np.median(df.revenue_mln)
MAD ,median
```

```python
def get_modified_z_score(x, median, MAD):
    return 0.6745*(x-median)/MAD
```

```python
get_modified_z_score(2787,median,MAD)
```

```python
df['modified_zscore'] = df.revenue_mln.apply(lambda x: get_modified_z_score(x, median, MAD))
df.head(3)
```

```python
df[df.modified_zscore>3.5]
```
